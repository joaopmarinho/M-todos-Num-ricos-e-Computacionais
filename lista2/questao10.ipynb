```python
# Simulation of motion of pendulum 
from scipy.integrate import odeint
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def model(x,t):
  y = x[0]
  dy = x[1]  
  K = 30
  xdot = [[],[]]
  xdot[0] = dy
  xdot[1] = -y + math.cos(t)
  return xdot
time = np.linspace(0,40,100)
z2 = odeint(model,[0,1],time)

# plot results
plt.plot(time,z2[:,0],'g:')
plt.plot(time,z2[:,1],'k-.')
plt.legend(['y (ODEINT)','dy/dt (ODEINT)'])
plt.xlabel('Time')
plt.show()
```
