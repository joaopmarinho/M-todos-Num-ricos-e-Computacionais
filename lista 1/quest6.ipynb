```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def model(y,t):
    dydt = (y+4)*np.cos(t)
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0,40, 200)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y,'b-',label=r'$\frac{dy}{dt}=(y+4) \; \cos(t)$')
plt.ylabel('y(t)')
plt.xlabel('t')
plt.legend(loc='best')
plt.show()
```

```python
def model2(y, t):
    dydt =  y * np.exp(t)
    return dydt

y0 = 1

t = np.linspace(0, 2)

y = odeint(model2, y0, t)

plt.plot(t, y, label = r'$\frac{dy}{dt}= y\; \exp(t)$')
plt.ylabel('y(t)')
plt.xlabel('t')
plt.legend(loc='best')
plt.show()
```
